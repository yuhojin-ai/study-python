```python
import seaborn as sns # 시각화 라이브러리 불러오기기

anscombe = sns.load_dataset("anscombe") # 예제 데이터셋 anscombe 로드
print(anscombe)
```

데이터의 통계량은 거의 같지만, 분포 형태는 완전히 다름
→ 통계 수치만 보고 분석하면 위험할 수 있다는 걸 보여주는 대표 예제

```python
import matplotlib.pyplot as plt

# lmplot()은 seaborn에서 선형회귀선이 포함된 산점도 그릴 때 사용
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", col_wrap=2, ci=None)
plt.suptitle("Anscombe's Quartet", y=1.05) # y: 그래프 보다 살짝 위
plt.show()

# data=anscombe | 시각화할 데이터프레임
# x="x", y="y" | x축, y축 데이터
# col="dataset" | dataset 컬럼을 기준으로 서브플롯을 나눔 (I, II, III, IV 각각 따로 그림)
# col_wrap=2 | 2열로 감싸서 보여줘서 2×2 형태로 만듦
# ci=None | 신뢰구간(confidence interval)을 표시하지 않음 (선만 깔끔하게)
```

```python
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", col_wrap=3, ci=None)
plt.suptitle("Anscombe's Quartet", y=1.05)
plt.show()
```

```python
sns.lmplot(data=anscombe, x="x", y="y", col="dataset", col_wrap=2)
plt.suptitle("Anscombe's Quartet", y=1.05)
plt.show()
```

## 04-3 matplotlib으로 그래프 그리기

```python
tips = sns.load_dataset("tips") # tips예제를 불러와 데이터프레임 반환
print(tips)
```

### 일변량 그래프 그리기

#### [실습] 히스토그램 그리기

```python
fig = plt.figure()

axes1 = fig.add_subplot(1, 1, 1)

axes1.hist(data=tips, x='total_bill', bins=10)

axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Frequency')

plt.show()
```

### 이변량 그래프 그리기

#### [실습] 산점도 그래프 그리기

```python
# scatter() 산점도 그래프프
# 두 변수의 관계를 보고 싶을 때	선형/비선형, 상관관계 여부 확인 가능
# 이상값(outlier) 찾을 때 눈에 띄는 점을 쉽게 발견 가능
# 군집 확인	점들이 모여 있는 형태를 확인 가능

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(tips['total_bill'], tips['tip'])

axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

plt.show()
print(type(axes1))
# color='red'	점 색상 지정
# marker='o'	마커(점 모양): 'o'=원, 'x', '^'=삼각형 등
# s=100	점 크기
# alpha=0.7	투명도 (0~1)
```

#### [실습] 박스 그래프 그리기

```python
print(tips)
```

```python
# 주식데이터에서 쓰임
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
    x=[
        tips[tips['sex'] == 'Female']['tip'],
        tips[tips['sex'] == 'Male']['tip']
    ],
    labels=['Female', 'Male']
)

axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot of Tips by Sex')

plt.show()
```

### 다변량 그래프 그리기

#### [실습] 변수가 여러 개인 그래프 그리기

```python
colors = {"Female": "#f1a340", "Male": "#998ec3"}

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(data=tips,
              x='total_bill',
              y='tip',
              s=tips['size']**2*10, # 점의 사이즈
              c=tips['sex'].map(colors), # 점의 크기
              alpha=0.5) # 화상도

axes1.set_title('Colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')

scatter_plot.suptitle('Total Bill vs Tip')

plt.show()
```

### 다양한 그래프 그려 보기

#### [실습] 일변량 그래프 그리기

##### 1. 히스토그램 그리기

```python
hist, ax = plt.subplots()

sns.histplot(data=tips, x='total_bill', ax=ax,)
ax.set_title('Total Bill Histogram')
plt.show()
print(type(ax))
```

##### 2. 밀도분포 그래프 그리기

```python
# 정규분포선
# 히스토그램과 비슷하지만, 매끄러운 곡선 형태로 데이터의 분포를 표현함
den, ax = plt.subplots()

sns.kdeplot(data=tips, x='total_bill', ax=ax)

ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')

plt.show()
```

```python
# 히스토그램과 KDE를 같이 그리기
sns.histplot(data=tips, x='total_bill', kde=True)
```

###### 3. 러그 그래프 그리기

```python
# 개별 데이터 포인트의 실제 위치를 확인
# 보조 정보로 많이 쓰임임
rug, ax = plt.subplots()

sns.rugplot(data=tips, x='total_bill', ax=ax)
sns.histplot(data=tips, x='total_bill', ax=ax)

ax.set_title('Rug Plot and Histogram of Total Bill')

plt.show()
```

```python
# kdeplot()이 전체 분포를 부드럽게 보여주고,
# rugplot()이 그 밑에 촘촘하게 선을 찍어서 개별 값의 위치

tips = sns.load_dataset("tips")

sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill", height=0.05)

plt.title("Total Bill Distribution with Rugplot")
plt.show()

# 주요 옵션션
# height	선분의 높이 (기본은 0.025)
# color	선분 색상
# axis='x' or 'y'	어느 축에 찍을지 설정
# hue	범주별로 색상을 나눌 수 있음 (seaborn 스타일)
```

##### 4. 분포 그래프 그리기

```python
# 정규 분포
# 히스토그램 + 밀토 + 러그플롯 한 번에 그림
fig = sns.displot(data=tips, x='total_bill', kde=True, rug=True)

fig.set_axis_labels(x_var='Total Bill', y_var='Count')
fig.figure.suptitle('Distribution of Total Bill')

plt.show()
```

```python
fig = sns.displot(data=tips, x='total_bill', kde=True, rug=True, hue='sex')

fig.set_axis_labels(x_var='Total Bill', y_var='Count')
fig.figure.suptitle('Distribution of Total Bill')

plt.show()
```

##### 5. 막대그래프 그리기

```python
count, ax = plt.subplots()

# 범주형 데이터의 개수(빈도) 를 세어서 막대그래프로 시각화하는 함수
sns.countplot(data=tips, x='day', palette='viridis', ax=ax)
# palette='viridis' 색상 팔레트 설정

ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')

plt.show()
```

### 이변량 그래프 그리기

##### 1. 산점도 그래프 그리기 ①

##### 2. 산점도 그래프 그리기 ②

##### 6. 막대그래프 그리기

```python
import numpy as np

bar, ax = plt.subplots()

sns.barplot(data=tips, x="time", y="total_bill", estimator=np.mean, ax=ax)

ax.set_title('Bar Plot of Average Total Bill for Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Average Total Bill')

plt.show()

# 자주 쓰느 estimator 함수
# np.mean	평균	기본값
# np.median	중앙값	이상치에 강함
# len 또는 lambda x: len(x)	빈도 수	그룹 내 데이터 개수
# np.std	표준편차	흔들림 크기 확인용
# np.max / np.min	최대/최소값	극단값 비교
```

```python
# relplot()은 관계형 그래프(산점도, 선 그래프 등)를 그리는 seaborn의 고급 시각화 함수
# relplot()은 산점도(scatterplot)나 선 그래프(lineplot)를 FacetGrid 구조로 자동 서브플롯까지 만들어주는 함수
# “같은 관계를 여러 조건으로 나눠서 보고 싶을 때”
anscombe_plot = sns.relplot(data=anscombe,
                            x="x",
                            y="y",
                            kind="scatter",
                            col="dataset", # dataset 값(I, II, III, IV)에 따라 그래프를 서브플롯으로 분할
                            col_wrap=2, # 	한 줄에 그래프 2개씩 정렬
                            height=2, # 각 subplot의 높이
                            aspect=1.6) # 가로 세로 비율 = 1.6 (그래프가 더 넓게 보임)

anscombe_plot.figure.set_tight_layout(True)

plt.show()
```

```python
seaborn_styles = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

fig = plt.figure()
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(data=tips, x="time", y="total_bill", ax=ax)
        violin.set_title(style)

fig.set_tight_layout(True)
plt.show()
```
