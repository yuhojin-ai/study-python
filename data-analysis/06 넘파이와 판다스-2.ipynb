```python
#ì½”ë“œ 6-1
import numpy as np

#ë¦¬ìŠ¤íŠ¸ë¥¼ ìƒì„±í•˜ê³  ë°°ì—´ë¡œ ë³€í™˜í•˜ê¸°
list1 = [1, 2, 3, 4]
a = np.array(list1)
print('a.shape: ', a.shape)
print('a[0]: ', a[0])
```

```python
# 1. array() - ë¦¬ìŠ¤íŠ¸ë¥¼ ë°°ì—´ë¡œ ë³€í™˜
a = np.array([1, 2, 3])
print("array:", a)

# 2. arange() - ì¼ì •í•œ ê°„ê²©ì˜ ìˆ˜ë¥¼ ndarray ë°°ì—´ë¡œ
b = np.arange(0, 10, 2)  # 0ë¶€í„° 10 ë¯¸ë§Œê¹Œì§€ 2 ê°„ê²©
print("arange:", b)

# 3. ones() - 1ë¡œ ì±„ìš´ ë°°ì—´ ìƒì„±
c = np.ones((2, 3))  # 2í–‰ 3ì—´
print("ones:\n", c)

# 4. zeros() - 0ìœ¼ë¡œ ì±„ìš´ ë°°ì—´ ìƒì„±
d = np.zeros((3, 2))  # 3í–‰ 2ì—´
print("zeros:\n", d)

# 5. empty() - ì´ˆê¸°í™”ë˜ì§€ ì•Šì€ ë¹ˆ ë°°ì—´ ìƒì„±
e = np.empty((2, 2))  # ê°’ì´ ì´ˆê¸°í™”ë˜ì§€ ì•ŠìŒ (ëœë¤ê°’)
print("empty:\n", e)

# 6. eye() ë˜ëŠ” identity() - ë‹¨ìœ„ í–‰ë ¬ ìƒì„±
f = np.eye(3)  # 3x3 ë‹¨ìœ„ í–‰ë ¬
print("eye:\n", f)

g = np.identity(3)  # 3x3 ë‹¨ìœ„ í–‰ë ¬
print("identity:\n", g)

# 7. linspace() - ì§€ì •í•œ êµ¬ê°„ì„ ê· ì¼í•˜ê²Œ ë‚˜ëˆˆ ê°’ ìƒì„±
h = np.linspace(0, 1, 5)  # 0ë¶€í„° 1ê¹Œì§€ 5ê°œì˜ ê· ë“±í•œ ê°’
print("linspace:", h)

# 8. full() - ì§€ì •í•œ ê°’ìœ¼ë¡œ ì±„ìš´ ë°°ì—´ ìƒì„±
i = np.full((2, 3), 7)  # 2í–‰ 3ì—´ ë°°ì—´ì„ 7ë¡œ ì±„ì›€
print("full:\n", i)
```

```python
#ì½”ë“œ 6-2
a = np.zeros(2)
print('a\n', a)
b = np.zeros((2,2))
print('b\n', b)
c = np.ones((2,3))
print('c\n', c)
d = np.full((2,3), 5)
print('d\n', d)
e = np.eye(3)
print('e\n', e)
```

### ğŸ“Œ ë„˜íŒŒì´ ë°°ì—´ ëª¨ì–‘ì„ ë³€ê²½í•˜ëŠ” ì£¼ìš” í•¨ìˆ˜/ë©”ì„œë“œ

|--------------------|----------------------------------------------------------------------|------------------------------------|
| `reshape(shape)`   | ë°°ì—´ì˜ í˜•íƒœë¥¼ ì§€ì •ëœ `shape`ìœ¼ë¡œ ë³€ê²½ (ì›ì†Œ ê°œìˆ˜ëŠ” ê°™ì•„ì•¼ í•¨)        | `a.reshape(2, 3)`                  |
| `flatten()`        | ë‹¤ì°¨ì› ë°°ì—´ì„ 1ì°¨ì›ìœ¼ë¡œ í‰íƒ„í™”í•¨ (ë³µì‚¬ë³¸ ë°˜í™˜)                       | `a.flatten()`                      |
| `ravel()`          | ë‹¤ì°¨ì› ë°°ì—´ì„ 1ì°¨ì›ìœ¼ë¡œ í‰íƒ„í™”í•¨ (ê°€ëŠ¥í•˜ë©´ ì›ë³¸ê³¼ ì—°ê²°ëœ ë·° ë°˜í™˜)    | `a.ravel()`                        |

| `resize(new_shape)`| ë°°ì—´ì„ ìƒˆë¡œìš´ í¬ê¸°ë¡œ ì§ì ‘ ë³€ê²½ (ì›ë³¸ ë°°ì—´ ìì²´ë¥¼ ìˆ˜ì •í•¨)             | `a.resize((3, 2))`                 |

```python
#ì½”ë“œ 6-5
#1ì°¨ì› ë°°ì—´ ìƒì„±í•˜ê¸°
a = np.arange(8)
print('a\n', a)

#ë‹¤ì°¨ì› ë°°ì—´ë¡œ ë³€ê²½í•˜ê¸°
a.shape = (2,4)
print('shape\n', a)

#1ì°¨ì› ë°°ì—´ë¡œ ë³€ê²½í•˜ê¸°
print('flatten\n', a.flatten())

#resize í•¨ìˆ˜ë¡œ ëª¨ì–‘ ë³€ê²½í•˜ê¸° (ì›ë³¸ ë°°ì—´ ìì²´ë¥¼ ìˆ˜ì •í•¨)
a.resize((4,2))
print('resize\n', a)
```

```python
print("reshape:\n", a.reshape(3, 2))
print("ravel:", a.ravel())
print("transpose:\n", a.transpose())
print("swapaxes:\n", a.swapaxes(0, 1))
print("expand_dims:", np.expand_dims(a, axis=0))  # (1, 2, 3)
print("squeeze:", np.squeeze(np.array([[[1, 2, 3]]])))  # -> [1 2 3]
```

```python
a = np.array([[1, 2], [3, 4]])

# flatten: ë³µì‚¬ë³¸ ìƒì„±
f = a.flatten()
f[0] = 100
print("flatten ê²°ê³¼:", f)      # [100   2   3   4]
print("ì›ë³¸ ë°°ì—´ a:", a)        # [[1 2]
                                #  [3 4]]  â†’ ì›ë³¸ ì˜í–¥ ì—†ìŒ

# ravel: ê°€ëŠ¥í•œ ê²½ìš° ë·° ë°˜í™˜
r = a.ravel()
r[0] = 200
print("ravel ê²°ê³¼:", r)        # [200   2   3   4]
print("ì›ë³¸ ë°°ì—´ a:", a)        # [[200 2]
                                #  [ 3  4]] â†’ ì›ë³¸ ì˜í–¥ ìˆìŒ
```

```python
#ì½”ë“œ 6-6
a = np.array([[0, 1, 2], [3, 4, 5]])
print('a\n', a)

b = a.transpose()
print('b\n', b)
c = a.T
print('c\n', c)
```

### **2. ë°°ì—´ ì›ì†Œ ì¡°íšŒ ë° ì •ë ¬**

```python
#ì½”ë“œ 6-7
# ë§ˆìŠ¤í‚¹: ì¡°ê±´ì— ë§ëŠ” ê°’ì„ ì¶œë ¥í•˜ëŠ” ê¸°ëŠ¥ëŠ¥
mask = np.array([0, 1, 1, 0], dtype=bool)
print(mask)

# randn(): í‰ê· ì´ 0, í‘œì¤€í¸ì°¨ê°€ 1ì¸ ì •ê·œë¶„í¬ë¥¼ ë”°ë¥´ëŠ” ë‚œìˆ˜ ìƒì„±ì„±
data = np.random.randn(4,2)
print('\ndata ì¶œë ¥\n',data) #ëœë¤ ë°ì´í„° ë°°ì—´ dataëŠ” ì±…ê³¼ ë‹¤ë¥´ê²Œ ë‚˜íƒ€ë‚¨

print('\në§ˆìŠ¤í‚¹ëœ ë°ì´í„° ì¶œë ¥\n',data[mask])
print('\në§ˆìŠ¤í‚¹ ì—­ì „ëœ ë°ì´í„° ì¶œë ¥\n',data[~mask])
```

```python
#ì½”ë“œ 6-8
posit = data[data > 0]
print('ì–‘ìˆ˜ ë°ì´í„° ì¶œë ¥', '\n',posit)

#ë‹¤ì¤‘ ì¡°ê±´
over1 = data[1][data[1] > 0]
print('ë‘ ë²ˆì§¸ í–‰ì˜ ì–‘ìˆ˜ ë°ì´í„° ì¶œë ¥','\n',over1)
```
