```python
# 정기권 고객의 시간대별 이용량을 구함
# df['count'] = 1
reg = df.loc[df['type'] == 'regular']
reg.groupby('datetime').agg({'count': 'sum'})
```

```python
# 정기권 고객의 시간대별 이용량을 구함
reg = df.loc[df['type'] == 'regular']
reg = reg.groupby('datetime').agg({'count': 'sum'}).reset_index()
reg.columns = ['datetime', 'regular']


cas = df.loc[df['type'] == 'casual']
cas = cas.groupby('datetime').agg({'count': 'sum'}).reset_index()
cas.columns = ['datetime', 'casual']

grp = df.loc[df['type'] == 'group']
grp = grp.groupby('datetime').agg({'count': 'sum'}).reset_index()
grp.columns = ['datetime', 'group']
```

```python
print(reg)
```

```python
print(cas)
```

```python
print(grp)
```

```python
print(reg.shape)
print(cas.shape)
print(grp.shape)

# 그룹으로 관리하면 총 row(data) 갯수가 확연하게 줄어 들어 관리가 용이함.
```

```python
rc = pd.merge(reg, cas, how='outer')
rcg = pd.merge(rc, grp, how='outer')
rcg
```

```python
df1 = pd.merge(reg, cas, how='inner')
df2 = pd.merge(rc, df1, how='inner')
df2
```

```python
rcg = rcg.fillna(0) # 결측값을 0으로 바꿈
rcg
```
