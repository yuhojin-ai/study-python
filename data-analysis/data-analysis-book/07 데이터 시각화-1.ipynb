```python
#코드 7-5
import numpy as np

# np.random.randn(): 평균이 0이고 표준편차가 1인 정규분표를 따르는 난수 생성
plt.plot(np.random.randn(10), 'k', label='one')         #검은색 실선
plt.plot(np.random.randn(10)*3, 'r--', label='two')     #빨간색 파선
plt.plot(np.random.randn(10)*10, 'g.', label='three')   #선 없이 초록색 점 마커로 표시
plt.legend() # 범례를 표시, plot()함수에서 label 지정한 것이 표시
plt.show()
```

```python
#코드 7-6
# alpha: 0에 가까울 수록 투명해 진다다
plt.subplot(2, 2, 1)                                    # 2행 2열 중 첫 번째 그래프
plt.plot(np.random.randn(10), 'b--')
plt.subplot(2, 2, 2)                                    # 2행 2열 중 두 번째 그래프
plt.plot(np.random.randn(100), 'r', alpha=0.7)
plt.subplot(2, 2, 3)                                    # 2행 2열 중 세 번째 그래프
plt.plot(np.random.randn(10), 'y^')
plt.subplot(2, 2, 4)                                    # 2행 2열 중 네 번째 그래프
plt.plot(np.random.randn(10), 'g.')
plt.show()
```

```python
#코드 7-12
!pip install networkx
import networkx as nx

G = nx.DiGraph()
G.add_nodes_from(['A','B','C','D','E'])
G.add_edges_from([('A','B'), ('B','C'), ('C','D')])

nx.draw(G, with_labels=True)
```

```python
#코드 7-13
#노드 추가 및 제거하기
g1 = nx.Graph()
g1.add_node('David') # 하나씩 추가
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy', 'Karen']) # 여러 개 추가

g1.remove_node('Julia')
#선 추가 및 제거하기
g1.add_edge('David', 'John')
g1.add_edge('David', 'Mary')
g1.add_edge('David', 'Karen')
g1.add_edges_from([('John', 'Mary'), ('John', 'Judy')])
g1.remove_edge('Mary', 'John')

nx.draw(g1, with_labels=True, font_weight='bold')
```

```python
#코드 7-14
d = dict(g1.degree)
nx.draw(g1, nodelist=d.keys(), node_size=[v * 1000 for v in d.values()],\
        with_labels=True, font_weight='bold')

g1.nodes
```

```python
#코드 7-15
g2 = nx.DiGraph()
g2.add_edges_from([(1, 2), (1, 3), (1, 4), (3, 4)])
nx.draw(g2, with_labels=True, font_weight='bold')

print(g2.in_degree) # 선이 들어오는 갯수
print(g2.out_degree) # 선이 나가는 갯수
```

```python
#코드 7-16
g3 = nx.DiGraph()
g3.add_weighted_edges_from([(1, 2, 10), (2, 3, 20)])
g3.add_edge(1, 3, weight=30)
pos = nx.spring_layout(g3) # 노드 간의 거리 연결을 자연스럽게 배치

nx.draw(g3, pos=pos, with_labels=True)
labels = nx.get_edge_attributes(g3, 'weight') # 딕셔너리 형태로 선들의 가중치 저장
nx.draw_networkx_edge_labels(g3, pos, edge_labels=labels);
```
