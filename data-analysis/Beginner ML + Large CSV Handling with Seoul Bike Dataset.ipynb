## LinearRegression 한눈에 보기

| 항목           | 설명                                |
|----------------|-------------------------------------|
|  목적         | 숫자 예측 (회귀 분석)               |
|  수학적 모델  | y = ax + b                          |
|  입력/출력    | 숫자 → 숫자                         |
|  핵심 개념    | 기울기(a), 절편(b)을 찾아 직선 생성 |
|  사용하는 곳  | 키 → 몸무게, 공부시간 → 점수 등     |



```python
from sklearn.linear_model import LinearRegression
import numpy as np

# x는 2차원 배열로 넣어야 해
x = np.array([[1], [2], [3], [4]])  # 공부 시간
y = np.array([50, 60, 65, 70])     # 시험 점수

model = LinearRegression()
model.fit(x, y)  # 학습

print("기울기:", model.coef_)     # a 값
print("절편:", model.intercept_)  # b 값

# 예측
print("5시간 공부하면 점수는?", model.predict([[5]]))
```

```python
model.score(X_test, y_test)
```

```python
import numpy as np
from sklearn.metrics import  mean_squared_error, r2_score
```

```python
y_pred = model.predict(X_test)
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(r2_score(y_test, y_pred))
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score

# 교차검증
score = cross_val_score(model, X_train, y_train, cv=10)
score
```
