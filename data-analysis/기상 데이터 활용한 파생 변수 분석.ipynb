# 불쾌지수의 범위: 쾌적함의 척도

* 20~29: 쾌적
* 30~39: 약간의 불쾌감
* 40~45: 큰 불쾌감
* 45 이상: 위험

<img src="불쾌지수.png">

<img src="humidex_calculate.png">

$$ H = T_{\text{air}} + 0.5555 \left\{ 6.11 \times \exp \left[ 5417.7530 \left( \frac{1}{273.15} - \frac{1}{273.15 + T_{\text{dew}}} \right) \right] - 10 \right\} $$

- where

  - $H$ denotes the Humidex
  - $T_{\text{air}}$ is the air temperature in °C
  - $T_{\text{dew}}$ is the dewpoint temperature in °C
  - $exp$ is the exponential function.

```python
Tair = 15
Tdew = 10

def humidex_calc(Tair, Tdew):
    inner = 5417.7530 * ( ( 1 / 273.15 ) - ( 1 / (273.15+Tdew)))
    inner = 6.11 * ( 2.71828 * inner) - 10
    return Tair + 0.5555 * inner
```

```python
humidex_calc(df['temp'], df['dew'])
```

```python
df['humidex'] = humidex_calc(df['temp'], df['dew'])
```

```python
df['discomfort'] = 0
df.loc[df['humidex'] >= 30, 'discomfort'] = 1 
df.loc[df['humidex'] >= 40, 'discomfort'] = 2
df.loc[df['humidex'] >= 46, 'discomfort'] = 3 
df.head()
```
